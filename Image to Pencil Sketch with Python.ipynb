## TASK 2 : Image to Pencil Sketch with Python
## LEVEL : Beginner


### Importing Required Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

### Reading the image in BGR format which by default

```python
img = cv2.imread("t&j.jpg")
plt.imshow(img)
```

### Converting BGR image to RGB format

```python
rgb_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

### Converting RGB image into GRAY_SCALE image

```python
gray_img = cv2.cvtColor(rgb_img,cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img)
```

### Inverting the GRAY image

```python
invt_img = cv2.bitwise_not(gray_img)
plt.imshow(invt_img)
```

### Blurring the image

```python
blur_img = cv2.GaussianBlur(invt_img,(21,21),sigmaX=0,sigmaY=0)
plt.imshow(blur_img)
```

### Inverting the blur_img

```python
invt_blur = cv2.bitwise_not(blur_img)
plt.imshow(invt_blur)
```

### Finally  we'll merge the gray_img and invt_blur, then converting that img_divide to a pencil sketch

```python
img_divide = cv2.divide(gray_img,invt_blur,scale=260)
skt = cv2.cvtColor(img_divide,cv2.COLOR_GRAY2BGR)
plt.imshow(skt)
plt.show()
```

### Saving the pencil sketch image as `.png`

```python
cv2.imwrite("Tom&Jerry.png",skt)
```
